```python
import pandas as pd

data = {
    'tweet': [
        "I love this product! üòç",
        "This is the worst thing ever...",
        "Meh, it‚Äôs okay I guess.",
        "Totally awesome! Would buy again.",
        "Terrible experience. Do not recommend.",
        "I'm not sure how I feel about this.",
        "Best day of my life!!",
        "Absolutely horrible. Waste of money.",
        "This was fine, nothing special.",
        "Wow! Just wow. Loved it!"
    ],
    'sentiment': [
        'positive',
        'negative',
        'neutral',
        'positive',
        'negative',
        'neutral',
        'positive',
        'negative',
        'neutral',
        'positive'
    ]
}

df = pd.DataFrame(data)
df.head()
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)  # Remove URLs
    text = re.sub(r'@\w+|#\w+', '', text)  # Mentions/Hashtags
    text = re.sub(r'[^A-Za-z\s]', '', text)  # Punctuation
    text = text.lower()
    tokens = text.split()  # Using split() instead of word_tokenize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return " ".join(tokens)

df['cleaned'] = df['tweet'].apply(clean_text)
df[['tweet', 'cleaned']]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Convert text into numerical vectors
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['cleaned'])
y = df['sentiment']
```

```python
from sklearn.model_selection import train_test_split

# Split into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```
